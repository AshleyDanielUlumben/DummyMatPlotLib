```python
import matplotlib.pyplot as plt
import csv
import pandas as pd
```

```python
#plotting a graph by reading a text file

with open("test.txt") as f:
    data = f.read()
#cutting after each line
data = data.split('\n')

#splitting the rows to get the x-axis and y-axis
x = [float(row.split(' ')[0]) for row in data]
y = [float(row.split(' ')[1]) for row in data]
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Practice Graph")
plt.show()
```

```python
#Just some practice graph

x_data = [1,2,3]
y_data = [2,4,1]
plt.plot(x_data, y_data)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Practice Graph")
plt.show()
```

```python
#Created a csv file from a text file
#the data row seems column

with open("Test2.txt") as f:
    lines = f.read()
#cutting after each line
lines = lines.split('\n')

#creating a new csv file
csv_Filename = "fdata3.csv"

with open(csv_Filename, mode = 'w', newline='') as csv_File:
    csv_Writer = csv.writer(csv_File)
    
    #write header
    header = lines[0].split(',')
    csv_Writer.writerow(header)
    
    #write data rows
    for line in lines[1:]:
        row = line.split(',')
        csv_Writer.writerow(row)
```

```python
#Trying out Pandas
#Line Graphs from a csv file

df = pd.read_csv('fdata2.csv', sep=',', parse_dates=True, index_col=0)
df.plot()
plt.show()
```
